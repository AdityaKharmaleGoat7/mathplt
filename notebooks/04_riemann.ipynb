```python
display(HTML("""
<div class='mp-anim-header'>
  <h2>1 &mdash; Zeros on the Critical Line</h2>
  <p>We evaluate &zeta;(&frac12;&nbsp;+&nbsp;it) for increasing values of <code>t</code>
     and plot the result as a point in the complex plane.
     <strong>Every time the curve passes through the origin, there is a zero</strong>
     of &zeta;(s) on the critical line.</p>
</div>
<div class='mp-watch'>
  <p><strong>Left panel</strong> &mdash; Re(&zeta;) and Im(&zeta;) vs t:
     zeros appear where both cross zero simultaneously, marked by dashed lines.<br>
  <strong>Right panel</strong> &mdash; path in the complex plane:
     each loop through the origin is one zero.<br>
  The Riemann Hypothesis predicts this curve will always originate from
  <code>Re(s) = &frac12;</code> &mdash; and all known zeros confirm this.</p>
</div>
"""))
```

```python
from mathplt.animations.riemann.zeros import RiemannZerosAnimator

config = AnimationConfig(fps=30, duration_seconds=20.0, figsize=(14, 6), dark_mode=True)
animator = RiemannZerosAnimator(config, t_min=0.5, t_max=50.0, n_zeros=15, resolution=2000)
AnimationWidget(animator).display()
```

```python
display(HTML("""
<div class='mp-anim-header'>
  <h2>2 &mdash; Critical Strip &mdash; Magnitude &amp; Phase</h2>
  <p>The <strong>critical strip</strong> is the region <code>0 &lt; Re(s) &lt; 1</code>
     where all nontrivial zeros live. We compute &zeta;(s) on a grid and show two heatmaps
     side by side.</p>
</div>
<div class='mp-watch'>
  <p><strong>Left &mdash; Magnitude</strong> <code>log(1 + |&zeta;(s)|)</code>:
     dark valleys along <code>Re(s) = &frac12;</code> are zeros.
     A bright spike at <code>s = 1</code> is the pole.<br>
  <strong>Right &mdash; Phase</strong> <code>arg(&zeta;(s))</code>:
     color encodes the output angle. Zeros appear where all colors meet at a point.<br>
  The Riemann Hypothesis is made <em>visible</em> here: all dark zero-spots
  line up on the <strong>vertical center line</strong> <code>Re(s) = &frac12;</code>.</p>
</div>
"""))
```

```python
from mathplt.animations.riemann.critical_strip import CriticalStripAnimator

config = AnimationConfig(fps=24, duration_seconds=15.0, figsize=(14, 7), dark_mode=True)
animator = CriticalStripAnimator(config, im_range=(0.5, 50.0), re_points=60, im_points=250)
AnimationWidget(animator).display()
```

```python
display(HTML("""
<div class='mp-anim-header'>
  <h2>3 &mdash; Zeta Surface &mdash; Domain Coloring</h2>
  <p><strong>Domain coloring</strong> of &zeta;(s) over the complex plane:
     hue encodes phase <code>arg(&zeta;(s))</code>,
     brightness rings encode <code>log|&zeta;(s)|</code> (each ring = &times;e in magnitude).</p>
</div>
<div class='mp-watch'>
  <p><strong>Nontrivial zeros</strong> &mdash;
     small dark pinch-points along the vertical line <code>Re(s) = &frac12;</code>,
     marked with white <code>+</code>.<br>
  <strong>Pole</strong> &mdash; bright chaotic region at <code>s = 1</code>
     where all colors meet and magnitude &rarr; &infin;.<br>
  <strong>Trivial zeros</strong> &mdash;
     dark spots along the negative real axis at <code>s = &minus;2, &minus;4, &minus;6, &hellip;</code><br>
  Count how many times the color wheel cycles around a feature to find its
  winding number (= zero/pole order).</p>
</div>
"""))
```

```python
from mathplt.animations.riemann.zeta_surface import ZetaSurfaceAnimator

config = AnimationConfig(fps=15, duration_seconds=8.0, figsize=(10, 10), dark_mode=True)
animator = ZetaSurfaceAnimator(
    config, re_range=(-1.5, 2.5), im_range=(0.5, 35.0),
    resolution=180, n_zeros_marked=10, animate_zoom=False,
)
AnimationWidget(animator).display()
```

```python
display(HTML("""
<div class='mp-anim-header'>
  <h2>4 &mdash; Winding Number &mdash; Argument Principle</h2>
  <p>The <strong>argument principle</strong> says: the number of zeros inside a closed contour C
     equals the number of times the image &zeta;(C) winds around the origin in the w-plane.
     This is a topological proof that doesn&rsquo;t require computing the zeros directly.</p>
</div>
<div class='mp-watch'>
  <p><strong>Left panel</strong> &mdash; the rectangular contour C in the s-plane,
     starting narrow and expanding upward to include more zeros.<br>
  <strong>Right panel</strong> &mdash; the image &zeta;(C) in the w-plane.
     Each time the contour swallows a new zero, the image curve
     <strong>loops around the origin once more</strong>.<br>
  The winding counter in the corner increments by <strong>1 per enclosed zero</strong>.
  Watch it match the count of known zeros in that imaginary range.</p>
</div>
"""))
```

```python
from mathplt.animations.riemann.winding_number import WindingNumberAnimator

config = AnimationConfig(fps=20, duration_seconds=20.0, figsize=(14, 7), dark_mode=True)
animator = WindingNumberAnimator(
    config, re_lo=0.1, re_hi=0.9, im_max=40.0, im_start=5.0, n_zeros=12
)
AnimationWidget(animator).display()
```

```python
display(HTML("""
<div class='mp-anim-header'>
  <h2>5 &mdash; Analytic Continuation</h2>
  <p>The Dirichlet series &sum;&nbsp;n<sup>&minus;s</sup> only converges for
     <code>Re(s) &gt; 1</code>. <strong>Analytic continuation</strong> extends &zeta;(s)
     to the entire complex plane while preserving all its complex-analytic properties.
     This animation makes that extension visible.</p>
</div>
<div class='mp-watch'>
  <p><strong>Left panel</strong> &mdash; domain coloring of the Dirichlet partial sum
     (first N terms), growing as the animation plays.<br>
  <strong>Right panel</strong> &mdash; the true &zeta;(s) computed via mpmath, always defined
     everywhere.<br>
  As N grows, the left panel <strong>converges to the right panel</strong> in the region
  <code>Re(s) &gt; 1</code>, but stays chaotic to the left &mdash; the Dirichlet series
  simply cannot converge there. The analytic continuation fills in what the series cannot reach.</p>
</div>
"""))
```

```python
from mathplt.animations.riemann.analytic_continuation import AnalyticContinuationAnimator

config = AnimationConfig(fps=15, duration_seconds=12.0, figsize=(14, 7), dark_mode=True)
animator = AnalyticContinuationAnimator(
    config, re_range=(-1.0, 3.0), im_range=(0.5, 25.0), resolution=120, max_terms=40
)
AnimationWidget(animator).display()
```

```python
display(HTML("""
<div class='mp-ref'>
  <h2>Mathematical Reference</h2>

  <h3 style='color:#a0aec0;margin:0 0 4px'>Definition (Dirichlet series, Re(s) &gt; 1)</h3>
  <pre>&zeta;(s) = &sum;<sub>n=1</sub><sup>&infin;</sup> 1/n<sup>s</sup></pre>

  <h3 style='color:#a0aec0;margin:12px 0 4px'>Euler product (connects to primes)</h3>
  <pre>&zeta;(s) = &prod;<sub>p prime</sub> 1 / (1 &minus; p<sup>&minus;s</sup>)</pre>
  <p>This means &zeta;(s) &ne; 0 for Re(s) &gt; 1 &mdash; no prime can make a product factor zero.</p>

  <h3 style='color:#a0aec0;margin:12px 0 4px'>Functional equation (used for analytic continuation)</h3>
  <pre>&zeta;(s) = 2<sup>s</sup> &pi;<sup>s&minus;1</sup> sin(&pi;s/2) &Gamma;(1&minus;s) &zeta;(1&minus;s)</pre>
  <p>Links &zeta;(s) to &zeta;(1&minus;s), letting us compute values with Re(s) &lt; 0
     from values with Re(s) &gt; 1.</p>

  <h3 style='color:#a0aec0;margin:12px 0 4px'>Zeros</h3>
  <p><strong>Trivial zeros:</strong> <code>s = &minus;2, &minus;4, &minus;6, &hellip;</code>
     &mdash; from the sin(&pi;s/2) factor going to zero.<br>
  <strong>Nontrivial zeros:</strong> all in the critical strip
  <code>0 &lt; Re(s) &lt; 1</code>. The Riemann Hypothesis: they all satisfy
  <code>Re(s) = &frac12;</code>.</p>

  <h3 style='color:#a0aec0;margin:12px 0 4px'>Connection to prime counting</h3>
  <p>The zeros of &zeta;(s) control the error in the Prime Number Theorem.
     If the Riemann Hypothesis is true, we get the sharpest estimate for
     how many primes &pi;(x) are below x:</p>
  <pre>&pi;(x) = Li(x) + O(&radic;x &middot; log x)</pre>
</div>
"""))
```
