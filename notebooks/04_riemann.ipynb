# Riemann Hypothesis — Visualizations

---

## What is the Riemann Zeta Function?

The **Riemann zeta function** `ζ(s)` is defined for complex numbers `s` with real part > 1 by:

$$\zeta(s) = \sum_{n=1}^{\infty} \frac{1}{n^s} = 1 + \frac{1}{2^s} + \frac{1}{3^s} + \frac{1}{4^s} + \cdots$$

It extends to the entire complex plane (except `s = 1`) via **analytic continuation**.  
This extension reveals the function's rich structure: a pole at `s = 1`, trivial zeros at negative even integers, and the mysterious **nontrivial zeros**.

---

## The Riemann Hypothesis

> *"All nontrivial zeros of ζ(s) have real part equal to 1/2."* — Bernhard Riemann, 1859

In other words: every nontrivial zero lies on the **critical line** `Re(s) = ½`.  
This is one of the **Millennium Prize Problems** — worth $1 million and unsolved for 165 years.

### Known zeros (first few imaginary parts `t` where `ζ(½ + it) = 0`)
```
14.135   21.022   25.011   30.425   32.935   37.586   40.919   43.327 …
```
All of these have real part exactly `½`. Over 10 trillion zeros have been computed — all on the critical line.

## 1. Zeros on the Critical Line

### How it works

We evaluate `ζ(½ + it)` for increasing values of `t` (the imaginary part).  
This gives a complex number for each `t`, which we plot as a point in the complex plane.

As `t` grows, the point traces a winding curve. **Every time the curve passes through the origin (0, 0), there is a zero of ζ(s)** on the critical line.

### What to watch
- The **left panel** shows `Re(ζ)` and `Im(ζ)` vs `t` — zeros appear where both cross zero simultaneously
- The **right panel** shows the path in the complex plane — zeros are visible as the curve looping through the origin
- Vertical dashed lines mark the **known zero locations**
- The magnitude `|ζ(½+it)|` dips to zero at each zero

````python
from mathplt.animations.riemann.zeros import RiemannZerosAnimator

config = AnimationConfig(
    fps=30,
    duration_seconds=20.0,
    figsize=(14, 6),
    dark_mode=True,
)
animator = RiemannZerosAnimator(
    config,
    t_min=0.5,
    t_max=50.0,
    n_zeros=15,
    resolution=2000,
)
AnimationWidget(animator).display()
````

## 2. Critical Strip — Magnitude & Phase

### How it works

The **critical strip** is the region `0 < Re(s) < 1` — this is where all the nontrivial zeros live.  
We compute `ζ(s)` on a grid covering this strip and display two heatmaps:

- **Left — Magnitude:** `log(1 + |ζ(s)|)`. Dark valleys along `Re(s) = ½` are zeros. The pole at `s = 1` would appear as a bright spike.
- **Right — Phase:** `arg(ζ(s))`. Color encodes the angle of the output. Zeros appear where the phase becomes undefined (all colors meeting at a point).

### What to watch
- Look for **dark spots** in the left panel — those are zeros of `ζ(s)`
- The zeros visibly align along the **vertical center line** `Re(s) = ½` — this is the Riemann Hypothesis made visible
- The right panel shows how the phase structure organizes around each zero

````python
from mathplt.animations.riemann.critical_strip import CriticalStripAnimator

config = AnimationConfig(fps=24, duration_seconds=15.0, figsize=(14, 7), dark_mode=True)
animator = CriticalStripAnimator(
    config,
    im_range=(0.5, 50.0),
    re_points=60,
    im_points=250,
)
AnimationWidget(animator).display()
````

## 3. Zeta Surface — Domain Coloring

### How it works

This uses **domain coloring** (same technique as the complex plane notebook) applied to `ζ(s)`:  
- **Hue** = phase `arg(ζ(s))`  
- **Brightness rings** = `log|ζ(s)|` — each ring marks one factor of `e` in magnitude

### What to look for

| Feature | Where to look |
|---|---|
| **Nontrivial zeros** | Small dark pinch-points along the vertical line `Re(s) = ½` — marked with white `+` |
| **Pole** | Bright chaotic region at `s = 1` — all colors meet, magnitude → ∞ |
| **Trivial zeros** | Dark spots along the negative real axis at `s = -2, -4, -6, …` |
| **Winding number** | Count how many times the color wheel cycles around a zero or pole |

The Riemann Hypothesis is visible here: all the dark zero-spots line up on `Re(s) = ½`.

````python
from mathplt.animations.riemann.zeta_surface import ZetaSurfaceAnimator

config = AnimationConfig(fps=15, duration_seconds=8.0, figsize=(10, 10), dark_mode=True)
animator = ZetaSurfaceAnimator(
    config,
    re_range=(-1.5, 2.5),
    im_range=(0.5, 35.0),
    resolution=180,
    n_zeros_marked=10,
    animate_zoom=False,
)
AnimationWidget(animator).display()
````

## 4. Winding Number — Argument Principle

### The core idea

The **argument principle** from complex analysis says:

$$N_{\text{zeros inside } C} = \frac{1}{2\pi} \Delta_C \arg(\zeta(s))$$

In plain language: **the number of zeros inside a closed contour C equals the number of times the image ζ(C) winds around the origin**.

### How this animation works

1. A rectangular contour `C` is drawn in the `s`-plane (left panel), starting narrow and expanding upward.
2. The image of `C` under `ζ` is drawn in the `w`-plane (right panel) — this is the curve `ζ(C)`.
3. As the contour grows to include more zeros, the image curve winds around the origin one additional time **per enclosed zero**.
4. The **winding count** is displayed — it should match the number of known zeros in that imaginary range.

### What to watch
- Each time the contour swallows a new zero, the image curve in the right panel **loops around the origin** once
- The winding count in the corner increments by 1 per zero
- This is a **topological proof** — it doesn't depend on computing the zeros directly

````python
from mathplt.animations.riemann.winding_number import WindingNumberAnimator

config = AnimationConfig(fps=20, duration_seconds=20.0, figsize=(14, 7), dark_mode=True)
animator = WindingNumberAnimator(
    config,
    re_lo=0.1,
    re_hi=0.9,
    im_max=40.0,
    im_start=5.0,
    n_zeros=12,
)
AnimationWidget(animator).display()
````

## 5. Analytic Continuation

### The problem

The Dirichlet series `Σ n^{-s}` only converges (gives a finite answer) when `Re(s) > 1`.  
But we need `ζ(s)` everywhere — especially at `s = ½` where the zeros are.

**Analytic continuation** is the process of extending a function beyond its original domain while preserving all its complex-analytic properties.

### How this animation works

The left panel shows the domain coloring of the **partial Dirichlet sum** (first N terms).  
The right panel shows the **true `ζ(s)`** computed via mpmath.

As N increases, the left panel:
- Converges to the true `ζ(s)` in the region `Re(s) > 1` (right half)
- Remains chaotic/wrong in `Re(s) ≤ 1` — the series simply doesn't converge there

### What to watch
- Watch the **right half** of the left panel converge to the right panel's colors as N grows
- The **left half** never stabilizes — this shows why analytic continuation is non-trivial
- The true `ζ(s)` (right panel) is smooth everywhere — the continuation "fills in" what the series can't reach

### The functional equation

The extension uses the functional equation:
$$\zeta(s) = 2^s \pi^{s-1} \sin\!\left(\frac{\pi s}{2}\right) \Gamma(1-s)\, \zeta(1-s)$$

This links `ζ(s)` to `ζ(1-s)`, letting us compute values with `Re(s) < 0` from values with `Re(s) > 1`.

````python
from mathplt.animations.riemann.analytic_continuation import AnalyticContinuationAnimator

config = AnimationConfig(fps=15, duration_seconds=12.0, figsize=(14, 7), dark_mode=True)
animator = AnalyticContinuationAnimator(
    config,
    re_range=(-1.0, 3.0),
    im_range=(0.5, 25.0),
    resolution=120,
    max_terms=40,
)
AnimationWidget(animator).display()
````

## Mathematical reference

**Definition (Dirichlet series, `Re(s) > 1`):**
$$\zeta(s) = \sum_{n=1}^{\infty} \frac{1}{n^s}$$

**Euler product (connects to primes):**
$$\zeta(s) = \prod_{p \text{ prime}} \frac{1}{1 - p^{-s}}$$

This means `ζ(s) ≠ 0` for `Re(s) > 1` — no prime can make a product factor zero.

**Functional equation (used for analytic continuation):**
$$\zeta(s) = 2^s \pi^{s-1} \sin\!\left(\frac{\pi s}{2}\right) \Gamma(1-s)\, \zeta(1-s)$$

**Trivial zeros:** `s = -2, -4, -6, …` — from the `sin(πs/2)` factor going to zero

**Nontrivial zeros:** All lie in the critical strip `0 < Re(s) < 1`.  
The Riemann Hypothesis claims they all satisfy `Re(s) = ½`.

**Connection to prime counting:**  
The zeros of `ζ(s)` control the error term in the Prime Number Theorem.  
If the Riemann Hypothesis is true, we get the sharpest possible estimate for how many primes are below a given number `x`:
$$\pi(x) = \text{Li}(x) + O(\sqrt{x}\log x)$$
