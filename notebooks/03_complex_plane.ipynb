```python
import matplotlib
matplotlib.rcParams['animation.embed_limit'] = 80

import ipywidgets as w
from IPython.display import display, HTML, clear_output
import manifold.animations
from manifold.animations.complex_plane import ComplexPlaneAnimator
from manifold.core.animator import AnimationConfig
from manifold.jupyter.widgets import AnimationWidget
from manifold.core.equation_parser import EquationParser

# ── Presets ───────────────────────────────────────────────────────────────────
PRESETS = [
    ("z\u00b2",          "z**2"),
    ("Cube roots",       "z**3 - 1"),
    ("Zeros & poles",    "(z**2 - 1) / (z**2 + 1)"),
    ("sin(z)",           "sin(z)"),
    ("exp(z)",           "exp(z)"),
    ("Cayley",           "(z - 1) / (z + 1)"),
    ("1/z",              "1 / z"),
    ("log(z)",           "log(z)"),
    ("Anim: z\u00b2+t",  "z**2 + t"),
    ("Anim: drift",      "(z - t) / (z + t)"),
]

style = {"description_width": "130px"}
lh = w.Layout(width="48%")

eq_input = w.Text(
    value="z**3 - 1",
    placeholder="e.g. z**3 - 1",
    description="f(z) =",
    style=style,
    layout=w.Layout(width="520px"),
)
eq_status = w.HTML(value='<span style="color:#68d391">&#10003; Valid</span>')

preset_buttons = []
for name, eq in PRESETS:
    btn = w.Button(description=name, layout=w.Layout(width="auto", margin="2px"))
    def _handler(_, e=eq): eq_input.value = e
    btn.on_click(_handler)
    preset_buttons.append(btn)

re_min  = w.FloatSlider(value=-2.5, min=-6,  max=-0.5, step=0.25, description="Re min",     style=style, layout=lh)
re_max  = w.FloatSlider(value=2.5,  min=0.5, max=6,    step=0.25, description="Re max",     style=style, layout=lh)
im_min  = w.FloatSlider(value=-2.5, min=-6,  max=-0.5, step=0.25, description="Im min",     style=style, layout=lh)
im_max  = w.FloatSlider(value=2.5,  min=0.5, max=6,    step=0.25, description="Im max",     style=style, layout=lh)
res_sl  = w.IntSlider(  value=300,  min=100, max=700,  step=50,   description="Resolution", style=style, layout=lh)
fps_sl  = w.IntSlider(  value=20,   min=8,   max=30,   step=2,    description="FPS",        style=style, layout=lh)
dur_sl  = w.FloatSlider(value=6,    min=2,   max=20,   step=1,    description="Duration (s)",style=style, layout=lh)
dark_cb = w.Checkbox(   value=True,  description="Dark mode",     style=style, layout=lh)

render_btn = w.Button(
    description=" Render",
    button_style="success",
    icon="play",
    layout=w.Layout(width="160px", height="40px"),
)
output = w.Output()

_parser = EquationParser()
def _validate(change=None):
    err = _parser.validate(eq_input.value, variables={"z", "t"})
    if err:
        eq_status.value = f'<span style="color:#fc8181">&#10007; {err}</span>'
        render_btn.disabled = True
    else:
        eq_status.value = '<span style="color:#68d391">&#10003; Valid</span>'
        render_btn.disabled = False
eq_input.observe(_validate, names="value")

def _render(_btn):
    import matplotlib.pyplot as plt
    with output:
        clear_output(wait=True)
        display(HTML('<span style="color:#a0aec0">Rendering &mdash; please wait&hellip;</span>'))
        plt.close("all")
        config = AnimationConfig(
            fps=fps_sl.value,
            duration_seconds=dur_sl.value,
            figsize=(8, 7),
            dpi=90,
            dark_mode=dark_cb.value,
        )
        animator = ComplexPlaneAnimator(
            config,
            equation=eq_input.value,
            re_range=(re_min.value, re_max.value),
            im_range=(im_min.value, im_max.value),
            resolution=res_sl.value,
        )
        clear_output(wait=True)
        AnimationWidget(animator).display()

render_btn.on_click(_render)

def _s(label): return w.HTML(f"<div class='mp-section'>{label}</div>")
def _h(text):  return w.HTML(f"<div class='mp-hint'>{text}</div>")
def _d():      return w.HTML("<hr class='mp-divider'>")

display(w.VBox([
    _s("Equation   f(z)"),
    w.HBox([eq_input, eq_status], layout=w.Layout(align_items="center")),
    _h("Variable: <code>z</code> (complex). Include <code>t</code> to animate. "
       "Functions: <code>sin cos exp log abs sqrt conj</code>. Power: <code>z**2</code>."),
    _d(),
    _s("Presets"),
    w.HBox(preset_buttons, layout=w.Layout(flex_flow="row wrap")),
    _d(),
    _s("Domain (complex plane window)"),
    w.HBox([re_min, re_max]),
    w.HBox([im_min, im_max]),
    _h("Zoom in near a zero or pole to see the local winding structure more clearly."),
    _d(),
    _s("Render"),
    w.HBox([res_sl, fps_sl]),
    w.HBox([dur_sl, dark_cb]),
    _h("Resolution 300 is fast for exploration; 500+ for final renders. "
       "FPS and Duration only affect animated equations using <code>t</code>."),
    w.HTML("<div style='height:10px'></div>"),
    render_btn,
    output,
], layout=w.Layout(padding="6px")))
```
