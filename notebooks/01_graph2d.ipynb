```python
import matplotlib
matplotlib.rcParams['animation.embed_limit'] = 50

import ipywidgets as w
from IPython.display import display, HTML, clear_output
import manifold.animations
from manifold.animations.graph2d import Graph2DAnimator
from manifold.core.animator import AnimationConfig
from manifold.jupyter.widgets import AnimationWidget
from manifold.core.equation_parser import EquationParser

# ── Presets ───────────────────────────────────────────────────────────────────
PRESETS = [
    ("Travelling wave",  "sin(x + t)"),
    ("Standing wave",    "sin(x) * cos(t)"),
    ("Gaussian wave",    "sin(x + t) * exp(-0.1 * x**2)"),
    ("Superposition",    "cos(3*x - 2*t) + 0.5 * sin(5*x + t)"),
    ("Frequency chirp",  "sin(x * t) / (1 + x**2)"),
    ("Gaussian pulse",   "exp(-0.05 * (x - 2*t)**2)"),
    ("Beats",            "sin(x + t) + sin(x + 1.05*t)"),
    ("Kink / soliton",   "tanh(x - 2*t)"),
    ("Decaying wave",    "sin(x + t) / (abs(x) + 1)"),
    ("Modulated wave",   "sin(5*x - 3*t) * sin(x - 0.3*t)"),
]

# ── Widgets ───────────────────────────────────────────────────────────────────
style = {"description_width": "130px"}
lh = w.Layout(width="48%")

eq_input = w.Text(
    value=PRESETS[0][1],
    placeholder="e.g. sin(x + t) * exp(-0.1 * x**2)",
    description="f(x, t) =",
    style=style,
    layout=w.Layout(width="580px"),
)
eq_status = w.HTML(value='<span style="color:#68d391">&#10003; Valid</span>')

preset_buttons = []
for name, eq in PRESETS:
    btn = w.Button(description=name, layout=w.Layout(width="auto", margin="2px"))
    def _handler(_, e=eq): eq_input.value = e
    btn.on_click(_handler)
    preset_buttons.append(btn)

x_min  = w.FloatSlider(value=-10, min=-30, max=-1, step=1,   description="x min",       style=style, layout=lh)
x_max  = w.FloatSlider(value=10,  min=1,   max=30, step=1,   description="x max",       style=style, layout=lh)
fps_sl = w.IntSlider(  value=30,  min=10,  max=60, step=5,   description="FPS",         style=style, layout=lh)
dur_sl = w.FloatSlider(value=10,  min=3,   max=30, step=1,   description="Duration (s)",style=style, layout=lh)
yclamp = w.Checkbox(   value=False, description="Fix y-axis limits", style=style, layout=lh)
y_min  = w.FloatSlider(value=-2,  min=-10, max=0,  step=0.5, description="y min",       style=style, layout=lh)
y_max  = w.FloatSlider(value=2,   min=0,   max=10, step=0.5, description="y max",       style=style, layout=lh)
dark_cb= w.Checkbox(   value=True, description="Dark mode",  style=style, layout=lh)

y_clamp_box = w.HBox([y_min, y_max])
y_clamp_box.layout.display = "none"
def _toggle_yclamp(change):
    y_clamp_box.layout.display = "" if change["new"] else "none"
yclamp.observe(_toggle_yclamp, names="value")

render_btn = w.Button(
    description=" Render Animation",
    button_style="success",
    icon="play",
    layout=w.Layout(width="210px", height="40px"),
)
output = w.Output()

_parser = EquationParser()

def _validate(change=None):
    err = _parser.validate(eq_input.value, variables={"x", "t"})
    if err:
        eq_status.value = f'<span style="color:#fc8181">&#10007; {err}</span>'
        render_btn.disabled = True
    else:
        eq_status.value = '<span style="color:#68d391">&#10003; Valid</span>'
        render_btn.disabled = False

eq_input.observe(_validate, names="value")

def _render(_btn):
    import matplotlib.pyplot as plt
    with output:
        clear_output(wait=True)
        display(HTML('<span style="color:#a0aec0">Rendering &mdash; please wait&hellip;</span>'))
        plt.close("all")
        config = AnimationConfig(
            fps=fps_sl.value,
            duration_seconds=dur_sl.value,
            figsize=(12, 5),
            dpi=90,
            dark_mode=dark_cb.value,
        )
        kwargs = dict(equation=eq_input.value, x_range=(x_min.value, x_max.value))
        if yclamp.value:
            kwargs["y_range"] = (y_min.value, y_max.value)
        animator = Graph2DAnimator(config, **kwargs)
        clear_output(wait=True)
        AnimationWidget(animator).display()

render_btn.on_click(_render)

def _s(label): return w.HTML(f"<div class='mp-section'>{label}</div>")
def _h(text):  return w.HTML(f"<div class='mp-hint'>{text}</div>")
def _d():      return w.HTML("<hr class='mp-divider'>")

display(w.VBox([
    _s("Equation   f(x, t)"),
    w.HBox([eq_input, eq_status], layout=w.Layout(align_items="center")),
    _h("Variables: <code>x</code> (position) and <code>t</code> (time). "
       "Functions: <code>sin cos tan exp log sqrt abs tanh</code>. Power: <code>x**2</code>."),
    _d(),
    _s("Presets"),
    w.HBox(preset_buttons, layout=w.Layout(flex_flow="row wrap")),
    _d(),
    _s("Domain"),
    w.HBox([x_min, x_max]),
    _d(),
    _s("Render"),
    w.HBox([fps_sl, dur_sl]),
    w.HBox([dark_cb, yclamp]),
    y_clamp_box,
    _h("Fix y-axis to prevent rescaling during decay or growth animations."),
    w.HTML("<div style='height:10px'></div>"),
    render_btn,
    output,
], layout=w.Layout(padding="6px")))
```
